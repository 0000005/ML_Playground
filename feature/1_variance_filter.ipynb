```python
import numpy as np
```

```python
# 生成标准的正太分布
a = np.random.randn(100)
print(np.var(a))
```

```python
#生成指定参数的正态分布。
b = np.random.normal(5,0.1,size=100)
print(np.var(b))
```

```python
# 构造特征向量，将两个向量合并为一个矩阵
X =np.vstack((a,b))
print(X.shape)
# 需要转置，将两行一百列转换为一百行两列。
X = X.T
print(X.shape)
```

```python
#过滤低方差的特征，这类特征意义不大。
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(0.1)
X_filtered = vt.fit_transform(X)
print(X_filtered)
print(X_filtered.shape)
```
