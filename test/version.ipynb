```python
import seaborn
import matplotlib
import pandas
import sklearn# 导入 Python 的内置模块 sys 和 platform
# 它们类似于 Java 中的 System 类，用于获取系统和运行时的底层信息
import sys
import platform


# 使用 sys 模块，获取更详细的系统版本信息（包含编译日期、编译器版本等）
print(f"Python 详细信息: \n{sys.version}")

# 打印包的内置版本号属性
# 这里的双下划线变量通常是 Python 模块的内置元数据
print("Seaborn 版本:", seaborn.__version__)
print("Matplotlib 版本:", matplotlib.__version__)
print("Pandas 版本:", pandas.__version__)
print("sklearn 版本:", sklearn.__version__)
```
